```python
# Define a function to get the vector representation of a class name
def get_vector(class_name):
    words = class_name.split()
    vector = np.zeros(w2v_model.vector_size)
    count = 0
    for word in words:
        if word in w2v_model.wv:
            vector += w2v_model.wv[word]
            count += 1
    if count != 0:
        vector /= count
    return vector
```
